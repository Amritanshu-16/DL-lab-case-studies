```python
# Case Study 1: MNIST Digit Recognition

# Objective:
# To classify handwritten digits (0вЂ“9) using CNN.
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import matplotlib.pyplot as plt
```

```python
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()
```

```python
# Normalize
x_train = x_train / 255.0
x_test = x_test / 255.0

# Reshape for CNN
x_train = x_train.reshape(-1, 28, 28, 1)
x_test = x_test.reshape(-1, 28, 28, 1)
```

```python
model = keras.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
history = model.fit(x_train, y_train, epochs=5, validation_data=(x_test, y_test))
```

```python
test_loss, test_acc = model.evaluate(x_test, y_test)
print("Test Accuracy:", test_acc)
```

```python
plt.imshow(x_test[0].reshape(28,28), cmap='gray')
pred = model.predict(x_test[0].reshape(1,28,28,1))
print("Predicted:", pred.argmax())
```

```python
# Case Study 2: CIFAR-10 CNN Classification

# Objective:
# To classify colored images into 10 categories using CNN.
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
(x_train, y_train), (x_test, y_test) = keras.datasets.cifar10.load_data()
```

```python
x_train = x_train / 255.0
x_test = x_test / 255.0
```

```python
model = keras.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(32,32,3)),
    layers.MaxPooling2D((2,2)),

    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),

    layers.Conv2D(64, (3,3), activation='relu'),

    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
history = model.fit(x_train, y_train, epochs=10, validation_data=(x_test, y_test))
```

```python
test_loss, test_acc = model.evaluate(x_test, y_test)
print("Accuracy:", test_acc)
```

```python
import numpy as np

class_names = ['airplane','automobile','bird','cat','deer','dog','frog','horse','ship','truck']

pred = model.predict(x_test[0].reshape(1,32,32,3))
print("Predicted:", class_names[np.argmax(pred)])
```
